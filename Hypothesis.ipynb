Plot means of scores obtained by two different methods

```python
import numpy as np

# Old method (population mean)
old_scores = np.array([70, 75, 72, 80, 78, 74, 76, 77, 73, 79])  # Example old scores

# New method (sample data)
new_scores = np.array([82, 78, 85, 79, 81, 76, 83, 79, 80, 84])  # Example new scores

# Calculate means
old_mean = np.mean(old_scores)
new_mean = np.mean(new_scores)

# Print means and standard deviations
print("Old method mean:", old_mean)
print("New method mean:", new_mean)
```

Calculate One Sample t-test T-Statistic and P-value for two different set of test scores

```python
import numpy as np
from scipy.stats import ttest_1samp

# Old method (population mean)
# old_scores = np.array([70, 75, 72, 80, 78, 74, 76, 77, 73, 79])  # Example old scores
# population_mean = np.mean(old_scores)  # Population mean

# New method (sample data)
new_scores = np.array([82, 78, 85, 79, 81, 76, 83, 79, 80, 84, 77, 81, 83, 79, 82, 80, 81, 77, 84, 79])  # Example new scores

# Perform one-sample t-test using manual calculation (formula)
new_mean_formula = np.sum(new_scores) / len(new_scores)
population_mean = new_mean_formula
new_std_dev_formula = np.sqrt(np.sum((new_scores - new_mean_formula)**2) / (len(new_scores) - 1))
t_statistic_manual = (new_mean_formula - population_mean) / (new_std_dev_formula / np.sqrt(len(new_scores)))

# Calculate p-value using the t-distribution CDF
degrees_of_freedom = len(new_scores) - 1
p_value_manual = 2 * (1 - stats.t.cdf(np.abs(t_statistic_manual), df=degrees_of_freedom))

# Perform one-sample t-test using scipy.stats.ttest_1samp
t_statistic_builtin, p_value_builtin = ttest_1samp(new_scores, population_mean)

# Print results
print("T-statistic (Manual calculation):", t_statistic_manual)
print("T-statistic (Built-in function):", t_statistic_builtin)
print("P-value (Manual calculation):", p_value_manual)
print("P-value (Built-in function):", p_value_builtin)
```

```python
import numpy as np
from scipy.stats import ttest_1samp

# Population parameters (Old method)
population_mean = 75  # Population mean score before the new teaching method was introduced
population_std_dev = 8

# Sample data (New method - test scores of 20 students taught using the new method)
test_scores = np.array([82, 78, 85, 79, 81, 76, 83, 79, 80, 84, 77, 81, 83, 79, 82, 80, 81, 77, 84, 79])

# Sample statistics
sample_mean = np.mean(test_scores)  # Sample mean score after implementing the new teaching method
sample_size = len(test_scores)

# Perform one-sample t-test using scipy.stats.ttest_1samp
t_statistic, p_value = ttest_1samp(test_scores, population_mean)

# Print results
print("Sample Mean:", sample_mean)
print("Sample Size:", sample_size)
print("Population Mean (Old method):", population_mean)
print("Population Standard Deviation:", population_std_dev)
print("T-statistic:", t_statistic)
print("P-value:", p_value)

# Interpretation
alpha = 0.05
if p_value < alpha:
    print("The p-value is less than the significance level.")
    print("Reject the null hypothesis. There is sufficient evidence to suggest that the new teaching method has led to a significant improvement in student performance compared to the population mean (old method).")
else:
    print("The p-value is greater than the significance level.")
    print("Fail to reject the null hypothesis. There is not enough evidence to conclude that the new teaching method has significantly affected student performance compared to the population mean (old method).")
```

```python
import numpy as np
from scipy import stats
from scipy.stats import ttest_ind
import matplotlib.pyplot as plt

num_samples = 100

# Simulated test scores for students taught using Method A
mean1 = 80
std_dev1 = 10
scores_method_1 = np.random.normal(mean1, std_dev1, num_samples)  # Mean: 80, Standard deviation: 10

# Simulated test scores for students taught using Method B
mean2 = 81
std_dev2 = 10
scores_method_2 = np.random.normal(mean2, std_dev2, num_samples)  # Mean: 85, Standard deviation: 10

# Calculate sample means
mean1 = np.mean(scores_method_1)
mean2= np.mean(scores_method_2)

# Calculate sample standard deviations
std_dev1 = np.std(scores_method_1, ddof=1)  # ddof=1 for unbiased estimator
std_dev2 = np.std(scores_method_2, ddof=1)


# Perform independent samples t-test
t_statistic, p_value = stats.ttest_ind(scores_method_1, scores_method_2)

# Define significance level
alpha = 0.05

# Print the results
print(f"T-statistic: {t_statistic}")
print(f"P-value: {p_value}")


# Plot the histograms
plt.hist(scores_method_1, bins=30, density=True, alpha=0.6, color='g', label=f'Mean={mean1}, StdDev={std_dev1}')
plt.hist(scores_method_2, bins=30, density=True, alpha=0.6, color='b', label=f'Mean={mean2}, StdDev={std_dev2}')
plt.xlabel('Value')
plt.ylabel('Probability')
plt.title('Histogram of Random Numbers with Different Means and Std Dev')
plt.legend()
plt.grid(True)
plt.show()

if p_value < alpha:
    print("YES, THERE IS SIGNIFICANT DIFFERENCE. Reject null hypothesis. There IS ENOUGH EVIDENCE to suggest a significant difference in test scores between Method A and Method B.")
else:
    print("NO, THERE IS NO SIGNIFICANT DIFFERENCE. Fail to reject null hypothesis. There is NOT ENOUGH EVIDENCE to conclude a significant difference in test scores between Method A and Method B.")

# Calculate pooled standard deviation
pooled_std_dev = np.sqrt(((len(scores_method_1) - 1) * std_dev1**2 +
                          (len(scores_method_2) - 1) * std_dev2**2) /
                          (len(scores_method_1) + len(scores_method_2) - 2))

# Calculate t-statistic using formula
t_statistic_formula = (mean1 - mean2) / (pooled_std_dev * np.sqrt(1 / len(scores_method_1) + 1 / len(scores_method_2)))

# Perform t-test using scipy.stats.ttest_ind
t_statistic_scipy, p_value_scipy = stats.ttest_ind(scores_method_1, scores_method_2)

# Calculate p-value using formula
degrees_of_freedom = len(scores_method_1) + len(scores_method_2) - 2
p_value_formula = 2 * (1 - stats.t.cdf(np.abs(t_statistic_formula), df=degrees_of_freedom))

# Print results
print("T-statistic (Formula):", t_statistic_formula); print("T-statistic (Scipy):", t_statistic_scipy)
print("P-value (Formula):", p_value_formula); print("P-value (Scipy):", p_value_scipy)
```
